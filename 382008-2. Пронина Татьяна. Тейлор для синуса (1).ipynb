```python
from IPython.core.display import display, HTML
from IPython.display import Image

print('Ряд Тейлора для функции sin x/n')
print('Необходимо написать функцию mysin(x, n) для вычисление суммы ряда Тейлора до n-ого ряда. Дополнительное условие: не должно быть циклов, следовательно, самые очевидные способы решения - циклы for и while - отметаются, остается только рекурсия.\nФункция синуса по Тейлору раскладывается так:')
display(Image(url='https://upload.wikimedia.org/math/3/3/4/3342704720a80cd858d7bedc5e4b7928.png', width = 400))
print('По сути, теперь у нас есть вся информация для написания функции. Приступим.\nПерепишем общий вид формулы для n-ого ряда в функцию:   (-1)**(n) * x**(2*n+1)/(2*n+1)!  ... и видим, что тут есть факториал. Придется и для него писать функцию.')
```

```python
def mysin(x,n):
    def fac(n):
        if n == 0:
            return 1
        return fac(n-1) * n
    if n==-1:
        return 0
    c=(-1) ** n * x**(2*n+1) / fac(2*n+1)
    return mysin(x,n-1)+c


print(mysin(2, 3))
```

```python
from math import pi, sin
import matplotlib.pyplot as plt
import numpy as np
def mysin(x,n):
    def fac(n):
        if n == 0:
            return 1
        return fac(n-1) * n
    if n==-1:
        return 0
    c=(-1) ** n * x**(2*n+1) / fac(2*n+1)
    return mysin(x,n-1)+c

x1 = pi/2
x2 = 11 * pi/2
x3 = 21 * pi/2
x4 = 31 * pi/2

def mysin_acc_plot(x):
    ns = range(1,60)
    deltas = list(map(lambda N: sin(x) - mysin(x,N), ns))
    plt.plot(ns, deltas)
    plt.xlabel('шаги')
    plt.ylabel('Дельты')
    plt.show()
mysin_acc_plot(x1)
mysin_acc_plot(x2)
mysin_acc_plot(x3)
mysin_acc_plot(x4)
```

```python
print("Как мы видим, точность страдает.\n\nМаксимальный по абсолютному значению член:")
```

```python
def mysin_max_el(x):
    results = [mysin(x,i) for i in range(80)]
    return max([abs(results[i]-results[i+1]) for i in range(len(results)-1)])

print("x1:",mysin_max_el(x1))
print("x2:",mysin_max_el(x2))
print("x3:",mysin_max_el(x3))
print("x4:",mysin_max_el(x4))
```
