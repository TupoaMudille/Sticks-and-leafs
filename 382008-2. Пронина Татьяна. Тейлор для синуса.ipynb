```python
def mysin(x,n):
    def fac(n):
        if n == 0:
            return 1
        return fac(n-1) * n
    if n==-1:
        return 0
    c=(-1) ** n * x**(2*n+1) / fac(2*n+1)
    return mysin(x,n-1)+c


print(mysin(2, 3))
```

```python
from math import pi, sin
import matplotlib.pyplot as plt
import numpy as np
def mysin(x,n):
    def fac(n):
        if n == 0:
            return 1
        return fac(n-1) * n
    if n==-1:
        return 0
    c=(-1) ** n * x**(2*n+1) / fac(2*n+1)
    return mysin(x,n-1)+c

x1 = pi/2
x2 = 11 * pi/2
x3 = 21 * pi/2
x4 = 31 * pi/2

def mysin_acc_plot(x):
    ns = range(1,60)
    deltas = list(map(lambda N: sin(x) - mysin(x,N), ns))
    plt.plot(ns, deltas)
    plt.show()
mysin_acc_plot(x1)
mysin_acc_plot(x2)
mysin_acc_plot(x3)
mysin_acc_plot(x4)
```

```python
from IPython.core.display import display, HTML
display(HTML("<font><p>Как мы видим, точность страдает.</p><p>Максимальный по абсолютному значению член:</p></font>"))
```

```python
def mysin_max_el(x):
    results = [mysin(x,i) for i in range(80)]
    return max([abs(results[i]-results[i+1]) for i in range(len(results)-1)])

print("x1:",mysin_max_el(x1))
print("x2:",mysin_max_el(x2))
print("x3:",mysin_max_el(x3))
print("x4:",mysin_max_el(x4))
```
